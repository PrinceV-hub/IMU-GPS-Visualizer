```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Data from CSV files
gps_file = "Location.csv"
accel_file = "Accelerometer.csv"
gyro_file = "Gyroscope.csv"

gps_data = pd.read_csv(gps_file)
accel_data = pd.read_csv(accel_file)
gyro_data = pd.read_csv(gyro_file)

# Standardize column names to lowercase and remove spaces
gps_data.columns = gps_data.columns.str.strip().str.lower()
accel_data.columns = accel_data.columns.str.strip().str.lower()
gyro_data.columns = gyro_data.columns.str.strip().str.lower()

# Plot GPS Trajectory
plt.figure(figsize=(10, 6))
sns.scatterplot(x=gps_data['longitude'], y=gps_data['latitude'], palette='coolwarm', edgecolor='k')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('GPS Trajectory')
plt.grid()
plt.show()

# Plot Accelerometer Data Separately
plt.figure(figsize=(12, 6))
plt.plot(accel_data['x'], label='Acc X', color='r')
plt.plot(accel_data['y'], label='Acc Y', color='g')
plt.plot(accel_data['z'], label='Acc Z', color='b')
plt.xlabel('Sample Index')
plt.ylabel('Acceleration (m/s²)')
plt.title('Accelerometer Data - X, Y, Z')
plt.legend()
plt.grid()
plt.show()

# Plot Gyroscope Data Separately
plt.figure(figsize=(12, 6))
plt.plot(gyro_data['x'], label='Gyro X', color='r')
plt.plot(gyro_data['y'], label='Gyro Y', color='g')
plt.plot(gyro_data['z'], label='Gyro Z', color='b')
plt.xlabel('Sample Index')
plt.ylabel('Angular Velocity (°/s)')
plt.title('Gyroscope Data - X, Y, Z')
plt.legend()
plt.grid()
plt.show()

# Combine X, Y, Z from Accelerometer and Gyroscope
plt.figure(figsize=(12, 6))
plt.plot(accel_data['x'], label='Acc X', linestyle='dashed', color='r')
plt.plot(gyro_data['x'], label='Gyro X', linestyle='solid', color='r')
plt.plot(accel_data['y'], label='Acc Y', linestyle='dashed', color='g')
plt.plot(gyro_data['y'], label='Gyro Y', linestyle='solid', color='g')
plt.plot(accel_data['z'], label='Acc Z', linestyle='dashed', color='b')
plt.plot(gyro_data['z'], label='Gyro Z', linestyle='solid', color='b')
plt.xlabel('Sample Index')
plt.ylabel('IMU Values')
plt.title('Combined Accelerometer and Gyroscope Data')
plt.legend()
plt.grid()
plt.show()
```
